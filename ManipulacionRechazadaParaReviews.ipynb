```python
import pandas as pd
```

```python
download_root = "https://raw.githubusercontent.com/CristianRiveraEsparza/MyBDCRE/main/"
airbnb = "AirBnB/reviews.csv"
df_reviews = pd.read_csv(download_root + airbnb)
```

```python
df_reviews.head(3)
```

```python
df_reviews.info()
```

```python
df_reviews['date']=pd.to_datetime(df_reviews['date'])
```

```python
df_reviews['date'].info()
```

```python
df_reviews_count=df_reviews
```

```python
df_reviews_count['year']=df_reviews_count['date'].dt.year
df_reviews_count['month']=df_reviews_count['date'].dt.month
df_reviews_count['day']=df_reviews_count['date'].dt.day
```

```python
df_reviews_count.head(3)
```

```python
df_reviews_count.groupby('listing_id')['date'].count()
```

```python
df_reviews_count=df_reviews_count.groupby('listing_id')['date'].count().reset_index()
```
